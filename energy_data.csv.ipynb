```python
import pandas as pd
import numpy as np

# Create date range
dates = pd.date_range(start="2025-01-01", periods=365)

zones = ["Zone1", "Zone2", "Zone3", "Zone4", "Zone5"]

data = []

for date in dates:
    for zone in zones:
        temp = np.random.normal(30, 5)
        humidity = np.random.uniform(40, 80)
        event = np.random.choice([0,1], p=[0.9,0.1])
        
        consumption = 200 + temp*5 + humidity*2 + event*50 + np.random.normal(0,10)
        
        data.append([date, zone, temp, humidity, event, consumption])

df = pd.DataFrame(data, columns=[
    "Date",
    "ZoneID",
    "AvgTemperature",
    "Humidity",
    "SpecialEvent",
    "EnergyConsumption"
])

df.to_csv("energy_data.csv", index=False)

print(df.head())
```
